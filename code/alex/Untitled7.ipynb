```python
url = r'http://archive.ceda.ac.uk/midas_stations/'
r = requests.get(url)
soup = BeautifulSoup(r.text,'html.parser')
```

```python
counties_soup = soup.find_all('select',{'name':'county'})
```

```python
counties_text = counties_soup[0].text[1:-1]
counties_list = []
while counties_text.find('\n') != -1:
    counties_list.append(counties_text[0:counties_text.find('\n')])
    counties_text = counties_text[counties_text.find('\n')+1:]
counties_list.append(counties_text)
```

```python
# =============================================================================
# Set up chromedriver
# =============================================================================
## THERE ARE 177 WEATHER STATIONS##
driver=webdriver.Chrome()


for i in counties_list:
    url = r'http://archive.ceda.ac.uk/cgi-bin/midas_stations/search_by_county.cgi.py?county='+i+'&minyear=&maxyear='
    driver.get(url)
    xpath = r'//*[@id="body"]/div[2]/div/div/form/center/input'
    driver.find_element_by_xpath(xpath).click()
```
